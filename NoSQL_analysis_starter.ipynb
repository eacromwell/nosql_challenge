### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {
    "scores.Hygiene": 20
}

results = list(establishments.find(query))
```

```python
# Use count_documents to display the number of documents in the result
count = db.establishments.count_documents(
    {
        "scores.Hygiene": 20
    }
)
print(f"There are {count} document records of restaurants with a hygiene score of 20.")
```

```python
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
```

```python
# Display the number of rows in the DataFrame
rows = df.shape[0]
print(f"There are {rows} rows in the DataFrame.")
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

regex_pattern = "Greenwich"

query = {
    "LocalAuthorityName": {
        "$regex": re.compile(regex_pattern, re.IGNORECASE)
    },
    "RatingValue": {"$gte": 4}
}

results = list(establishments.find(query))
```

```python
# Use count_documents to display the number of documents in the result

count = db.establishments.count_documents(
 {
    "LocalAuthorityName": {
        "$regex": re.compile(regex_pattern, re.IGNORECASE)
    },
    "RatingValue": {"$gte": 4}
}
)
print(f"There are {count} document records of restaurants with a hygiene score of 20.")
```

```python
# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame

rows = df.shape[0]
print(f"There are {rows} rows in the DataFrame.")
```

```python
# Export a CSV
df.to_csv('Resources/data_2_london_rating_4_and_above.csv', header=True)

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Create variables of the query. This can be modified easily.

businessname = "Penang Flavours"
document = db.establishments.find_one({"BusinessName": businessname})
latitude = document['geocode']['latitude']
longitude = document['geocode']['longitude']


print(f"The latitude is {latitude} and the longitude is {longitude} for {businessname}.")
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5

degree_search = 0.01

# Create a query
query = {
    "geocode.latitude": {"$gte": latitude - degree_search, "$lte": latitude + degree_search},
    "geocode.longitude": {"$gte": longitude - degree_search, "$lte": longitude + degree_search},
    "RatingValue": 5
}

# Sort by hygiene score
sort = [("scores.Hygiene", pymongo.DESCENDING)]

# Retrieve the results
results = list(db.establishments.find(query).sort(sort))

# Print the first result for viewing
pprint(results[0])
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame

# Export a CSV
df.to_csv('Resources/data_3_top_5_rating_hygiene_near_penang.csv', header=True)

rows = df.shape[0]
print(f"There are {rows} rows in the DataFrame.")
```

```python
# Display the first 10 Rows of the DataFrame
df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Print the number of documents in the result

results = list(db.establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"There are {len(results)} documents in the record.")
```

```python
# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame

rows = df.shape[0]
print(f"There are {rows} rows in the DataFrame.")

# Export a CSV
df.to_csv('Resources/data_4_hygiene_local_authority_sorted.csv', header=True)

# Display the first 10 Rows of the DataFrame
df.head(10)
```
